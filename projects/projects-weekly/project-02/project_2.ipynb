```python
#Number of weeks a song stayed in top 10 in the billboard

def weeksInTopTen(df):
    
     dfMap = {}
    
     for r in range(0, 317):
        
        song = df.iloc[r,2]
        
        df1 = df.iloc[r, 7:]
        
        df1 = pd.to_numeric(df1, errors= 'coerce')
        #print df1
        #print type(df1)
    
        N1 = df1[(df1<10) & (df1>0)].count()    
        #print r, song, N1
        dfMap[song] = (r, N1)
        
     df3 = pd.DataFrame.from_dict(dfMap, orient = 'index')
     df3.columns = ['song_index', 'NumOfWksInTopTen']
     df3 = df3.sort_values(by = 'NumOfWksInTopTen', ascending = False)
    
     print df3
    
weeksInTopTen(df)


```

"Breathe", "Maria, Maria", "Kryptonite" and "Everything You Want" are among the songs that have been for most weeks in the top ten position in bill board.

```python
#Number of weeks a song stayed in billboard

def weeksInBillBoard (df):
    
     dfMap = {}
    
     for r in range(0, 317):
        
              
        song = df.iloc[r,2]
        
        df1 = df.iloc[r, 7:]
        
        df1 = pd.to_numeric(df1, errors= 'coerce')
        #print df1
        #print type(df1)
    
        N1 = df1[(df1<101) & (df1>0)].count()    
        #print r, song, N1
        dfMap[song] = (r, N1)
        
     df4 = pd.DataFrame.from_dict(dfMap, orient = 'index')
     df4.columns = ['song_index', 'NumOfWksInBillBoard']
     df4 = df4.sort_values(by = 'NumOfWksInBillBoard', ascending = False)
     print "Average number of weeks in billboard for a song", df4['NumOfWksInBillBoard'].mean()
     print df4
    
     
               
weeksInBillBoard(df)
```

"Higher", "Amazed", "Kryptonite" and "Breathe" are among the songs that have been in the billboard for the longest.

```python
# Testing the null hypothesis

from scipy import stats

def weeksInBillBoard_G_B_G(df):
    
     dfMap = {}
    
     for r in range(0, 317):
        
        song = df.iloc[r,2]      
        genre = df.iloc[r,4]
        
        sr1 = df.iloc[r, 7:]
        
        sr1 = pd.to_numeric(sr1, errors= 'coerce')
    
        N1 = sr1[(sr1<101) & (sr1>0)].count()    
        #print r, song, N1
        
        dfMap[song] = (genre, N1)
        
     df4 = pd.DataFrame.from_dict(dfMap, orient = 'index')
     df4.columns = ['genre', 'NumOfWksInBillBoard']
     df4 = df4.sort_values(by = 'genre')
     #print "Average number of weeks in billboard", df4['NumOfWksInBillBoard'].mean()
     #print df4
     
     RDF= df4[df4['genre']=='Rock'][['NumOfWksInBillBoard']].reset_index(drop= True) 
     #print RDF
    
     A1 = RDF['NumOfWksInBillBoard']
     
     CDF= df4[df4['genre']=='Country'][['NumOfWksInBillBoard']].reset_index(drop= True) 
     #print CDF 
    
     C1 = CDF['NumOfWksInBillBoard']
        
     Hypo = stats.ttest_ind(A1, C1)
     print Hypo
    
weeksInBillBoard_G_B_G(df)
```

Null Hypothesis: The average number of weeks a Rock music stays in bill board is equal to the average number of weeks a Country music stays in bill board. Since pvalue = 0.4 the null hypothesis is true.
