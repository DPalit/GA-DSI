```python
import pandas_datareader.data as web
import datetime
```

```python
start = datetime.datetime(2014, 12, 01)
end = datetime.datetime(2016, 12, 1)
```

```python
goog = web.DataReader('FB', 'google', start, end)
```

```python
print goog
```

```python
goog['2014']['Close'].mean()
```

```python
goog[datetime.datetime(2015, 5, 1): datetime.datetime(2015, 8, 1)]['Close'].median()
```

```python
goog['2016']['Close'].min()
```

```python
goog['2016']['Close'].max()
```

```python
goog[['Close']].resample('W').max()
```

```python
goog.resample('D')[['Close']].max()
```

```python
goog.resample('W', label='left')[['Close']].max()
```
